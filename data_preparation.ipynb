```python
h5py_file =output_file[0] 
with h5py.File(h5py_file, 'r') as h5_file:
  keys = list(h5_file.keys())
  selected_key = keys[2]
  selected_data = h5_file[selected_key][:]
  
selected_key        
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate a sample signal
# signal = np.sin(50 * 2 * np.pi * t)  # Example sinusoidal signal
signal=selected_data[:,2]
# signal.shape
# Define scales for CWT
scales = np.arange(100, 128)

# Perform Continuous Wavelet Transform
cwt_coefficients, frequencies = pywt.cwt(signal, scales, 'morl')

# Plot the CWT coefficients
plt.imshow(np.abs(cwt_coefficients), extent=[0, 60,100, 128], cmap='jet', aspect='auto')
plt.colorbar(label='Magnitude')
plt.xlabel('Time')
plt.ylabel('Scale')
plt.title('Continuous Wavelet Transform')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft

# Generate a sample signal
fs = 1000  # Sampling frequency
t = np.linspace(0, 5, 5 * fs, endpoint=False)  # Time vector
f1 = 10  # Frequency of the first component
f2 = 200  # Frequency of the second component
signal = selected_data[:,2]

# Compute the STFT
frequencies, time_points, magnitude = stft(signal, fs, window='hann', nperseg=256, noverlap=128)

# Plot the magnitude spectrogram
plt.figure(figsize=(10, 6))]
plt.pcolormesh(time_points, frequencies, np.abs(magnitude), shading='auto')
plt.colorbar(label='Magnitude')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('STFT Magnitude Spectrogram')
plt.ylim(0, 300)  # Set the frequency range for visualization
plt.show()
```
